```python
import keyboard
import uuid
import time
from PIL import Image
from mss import mss
```

```python
#top,left,width,height are set according to where the program see necessary part of screen.
mon = {"top":473, "left":817, "width":280, "height":100}
sct = mss()

i = 0

def record_screen(record_id, key):
    global i
    i+=1
    print(f'{key}: {i}')
    
    img = sct.grab(mon)
    im = Image.frombytes("RGB",img.size,img.rgb)
    im.save(f'./trackresimleri/{key}_{record_id}_{i}.png')
    
is_exit = False

def exit():
    global is_exit
    is_exit = True
    
keyboard.add_hotkey("esc",exit)

record_id = uuid.uuid4()

while True:
    if is_exit:
        break
    
    try:
        if keyboard.is_pressed(keyboard.KEY_UP):
            record_screen(record_id,"up")
            time.sleep(0.1)
        elif keyboard.is_pressed(keyboard.KEY_DOWN):
            record_screen(record_id,"down")
            time.sleep(0.1)
        elif keyboard.is_pressed("right"):
            record_screen(record_id,"right")
            time.sleep(0.1)
    except RuntimeError:
        continue
        
        
```
