```python
import cv2
import matplotlib.pyplot as plt
```

```python
albert = cv2.imread("albert.jpg",0)
plt.figure(),plt.imshow(albert,cmap="gray"),plt.axis("off")
```

```python
#sınıflandırıcı
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
face_rect = face_cascade.detectMultiScale(albert)
```

```python
for (x,y,w,h) in face_rect:
    cv2.rectangle(albert,(x,y),(x+w,y+h),(255,255,255),10),plt.axis("off")
plt.figure(),plt.imshow(albert,cmap="gray"),plt.axis("off")
```

```python
madrid = cv2.imread("realmadrid.jpg",0)
plt.figure(),plt.imshow(madrid,cmap="gray"),plt.axis("off")
```

```python
face_rect = face_cascade.detectMultiScale(madrid,minNeighbors = 15)
#minNeighbors = kutucukların yanında 15 kutucuk varsa göster diyor.
#0 yapıldığında gerçek yüzler birden çok algılanıyor zaten, bu da işe yaramasını sağlıyor
for (x,y,w,h) in face_rect:
    cv2.rectangle(madrid,(x,y),(x+w,y+h),(255,255,255),10),plt.axis("off")
plt.figure(),plt.imshow(madrid,cmap="gray"),plt.axis("off")
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    
    if ret:
        face_rect = face_cascade.detectMultiScale(frame,minNeighbors = 7)
        
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,255),10)
        cv2.imshow("face detect",frame)
    if cv2.waitKey(1) & 0xFF == ord("q"): break

cap.release()
cv2.destroyAllWindows()
```
