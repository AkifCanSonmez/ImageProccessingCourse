```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#resmi ie aktar
img = cv2.imread("london.jpg",0)
plt.figure(),plt.imshow(img,cmap="gray"),plt.axis("off"),plt.title("Orijinal")
```

```python
edges = cv2.Canny(image=img, threshold1 = 0, threshold2 = 255)
plt.figure(),plt.imshow(edges,cmap="gray"),plt.axis("off"),plt.title("Edges")
```

```python
med_val = np.median(img)
print(med_val)
```

```python
low = int(max(0,(1-0.33)*med_val))
high = int(min(255,(1+0.33)*med_val))
print(low,high)
```

```python
edges = cv2.Canny(image=img,threshold1 = low, threshold2 = high)
plt.figure(),plt.imshow(edges, cmap = "gray"),plt.axis("off"),plt.title("Edges with Low-High")
```

### Su kenarlar hala alglanyor, resme bluring yaparak zm yaplacak

```python
#blur
blurred_img = cv2.blur(img,ksize=(5,5))
plt.figure(),plt.imshow(blurred_img,cmap="gray"),plt.axis("off")

med_val = np.median(blurred_img)
print(med_val)

low = int(max(0,(1-0.33)*med_val))
high = int(min(255,(1+0.33)*med_val))
print(low,high)
```

```python
edges = cv2.Canny(image=blurred_img,threshold1 = low, threshold2 = high)
plt.figure(),plt.imshow(edges, cmap = "gray"),plt.axis("off"),plt.title("Edges with Blurred_img")
```
