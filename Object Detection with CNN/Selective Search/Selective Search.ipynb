## Piramit Gösterimi ve Kayan Pencereden Daha Etkili Bir Yöntemdir

```python
import cv2
import random
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Resimler/pyramid.jpg")
image = cv2.resize(image,dsize=(600,600))
cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### İlklendirme


```python
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
ss.setBaseImage(image)
ss.switchToSelectiveSearchQuality()
```

```python
print("start")
rects = ss.process()
```

```python
output = image.copy()

for (x,y,w,h) in rects[:50]:
    color = [random.randint(0,255) for j in range(0,3)]
    cv2.rectangle(output,(x,y),(x+w, y+h),color,2)


cv2.imshow("output",output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(),plt.imshow(output),plt.axis("off")
```
